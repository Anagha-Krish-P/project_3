```python
num_cols = ['price', 'area', 'bedrooms', 'bathrooms', 'stories', 'parking']
plt.figure(figsize=(18, 12))

for i, col in enumerate(num_cols):
    plt.subplot(3, 3, i + 1)
    plt.hist(df[col].dropna(), bins=20, color='teal', edgecolor='blue')
    plt.title(f'{col} Distribution')
    plt.xlabel(col)
    plt.ylabel('Count')

plt.tight_layout()
plt.show()
```

```python

plt.figure(figsize=(10, 5))

# First subplot - Age
plt.subplot(1, 2, 1)
sns.boxplot(y=df['price'], color='#a93226')
plt.title('Boxplot of price')

# Second subplot - Fare
plt.subplot(1, 2, 2)
sns.boxplot(y=df['area'], color='#b03a2e')
plt.title('Boxplot of area')

plt.tight_layout()
plt.show()
```

```python
cols = ['area', 'price']

for col in cols:
    # Calculate IQR
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    iqr = q3 - q1
    lower = q1 - 1.5 * iqr
    upper = q3 + 1.5 * iqr

    # Replace outliers with the mean of the column
    mean_val = df[col].mean()
    df[col] = df[col].apply(lambda x: mean_val if x < lower or x > upper else x)

    # Plot boxplot
    sns.boxplot(x=df[col])
    plt.title(f'Boxplot of {col.capitalize()}')
    plt.show()
```

```python
x=df.drop(columns='price')
x
```

```python
y = df.iloc[:, 0]
y
```

```python
df['price'].value_counts()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)# 80% training, 20% testing

x_train
```

```python
print("Training set size:", x_train.shape)
print("Test set size:", x_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()

model.fit(x_train, y_train)
y_pred = model.predict(x_test)


```

```python

print("Intercept:", model.intercept_)

print("Coefficients:")
for feature, coef in zip(x.columns, model.coef_):
    print(f"{feature}: {coef:.2f}")
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Calculate metrics
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

# Print results
print(f"Mean Absolute Error (MAE): {mae:.2f}")
print(f"Mean Squared Error (MSE): {mse:.2f}")
print(f"RÂ² Score: {r2:.4f}")
```

```python
plt.figure(figsize=(8, 5))
plt.scatter(x_test['area'], y_test, color='skyblue', label='Actual Price')
plt.plot(x_test['area'], model.predict(x_test), color='teal', label='Predicted Price Line')

plt.xlabel('Area (in sq ft)')
plt.ylabel('Price')
plt.title('Linear Regression: Area vs Price')
plt.legend()
plt.grid(True)
plt.show()
```
