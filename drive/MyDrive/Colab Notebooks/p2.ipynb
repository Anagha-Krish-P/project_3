```python
df.columns = df.columns.str.strip()  # Removes tabs, spaces, etc.
```

```python
df['Age'] = df['Age'].fillna(df['Age'].mean())
```

```python
# Fill missing values with the mode (most frequent value)
df['Cabin'] = df['Cabin'].fillna(df['Cabin'].mode()[0])
```

```python
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
lst=['Name','Sex','Ticket','Cabin','Embarked']  #each unique cateogary is assign to unique integer value
for i in lst:
  df[i]=le.fit_transform(df[i])
  df
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df_normalized = df.copy()
df_normalized[df.columns] = scaler.fit_transform(df)

print(df_normalized)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df_standardized = df.copy()
df_standardized[df.columns] = scaler.fit_transform(df)

print(df_standardized)
```

# ***visualizing numerical variable ***

```python
num_cols = ['PassengerId', 'Survived', 'Pclass', 'Age', 'SibSp', 'Parch', 'Fare']
```

```python
plt.figure(figsize=(18, 12))  # Adjust size as needed

# Loop through each numerical column and plot histogram
for i, col in enumerate(num_cols):
    plt.subplot(3, 3, i + 1)  # 3 rows, 3 columns of plots
    plt.hist(df[col].dropna(), bins=20, color='#598dba', edgecolor='black')
    plt.title(f'{col} Distribution')
    plt.xlabel(col)
    plt.ylabel('Count')

plt.tight_layout()
plt.show()
```

# *continuous variable visual representation *

```python
plt.figure(figsize=(12, 6))

# First subplot - Age
plt.subplot(1, 2, 1)
sns.boxplot(y=df['Age'], color='#a93226')
plt.title('Boxplot of Age')

# Second subplot - Fare
plt.subplot(1, 2, 2)
sns.boxplot(y=df['Fare'], color='#b03a2e')
plt.title('Boxplot of Fare')

plt.tight_layout()
plt.show()
```

```python
q1=df['Age'].quantile(0.25)
q3=df['Age'].quantile(0.75)
iqr=q3-q1
lower=q1-(iqr*1.5)
upper=q3+(iqr*1.5)
for i in df['Age']:
  if i<lower or i>upper:
    df['Age']=df['Age'].replace(i,np.mean(df['Age']))
# This line was previously incorrectly indented inside the for loop
sns.boxplot(df['Age'])
```

```python
q1=df['Fare'].quantile(0.25)
q3=df['Fare'].quantile(0.75)
iqr=q3-q1
lower=q1-(iqr*1.5)
upper=q3+(iqr*1.5)
for i in df['Fare']:
  if i<lower or i>upper:
    df['Fare']=df['Fare'].replace(i,np.mean(df['Fare']))
# This line was previously incorrectly indented inside the for loop
sns.boxplot(df['Fare'])
```
